```python
import pandas as pd
import numpy as np
from pathlib import Path
import re

archivos = list(Path("DATOS_TFM").rglob("*.xls*"))

def extraer_metadata(nombre):
    nombre = nombre.upper()

    anio = re.search(r"(20\d{2})", nombre)
    anio = int(anio.group(1)) if anio else None

    if "INS" in nombre:
        organismo = "INSS"
    elif "MUT" in nombre:
        organismo = "Mutuas"
    elif "ISM" in nombre:
        organismo = "ISM"
    elif "SIS" in nombre:
        organismo = "Sistema"
    else:
        organismo = "Otro"

    if "COMUNES" in nombre:
        contingencia = "Comunes"
    elif "PROFESIO" in nombre:
        contingencia = "Profesionales"
    elif "PROPIA" in nombre:
        contingencia = "Cuenta propia"
    else:
        contingencia = "Otro"

    if "HOM" in nombre:
        sexo = "Hombre"
    elif "MUJ" in nombre:
        sexo = "Mujer"
    else:
        sexo = "Total"

    if "DUR" in nombre:
        tipo = "DUR"
    elif "PRE" in nombre:
        tipo = "PRE"
    else:
        tipo = "OTRO"

    return anio, organismo, contingencia, sexo, tipo


def leer_dur(archivo):
    df = pd.read_excel(archivo, header=4)
    df = df.rename(columns={
        "AMBITO GEOGRAFICO": "Provincia",
        "ÁMBITO GEOGRÁFICO": "Provincia",
        "Duración media de los procesos con alta en el periodo": "Duracion_media",
        "Duración media de los procesos finalizados en el periodo": "Duracion_media",
        "Nº de procesos cuya alta se ha producido en el periodo": "Procesos_alta",
        "Nº de días de baja de procesos cuya alta se ha producido en el periodo": "Dias_baja",
        "Nº de días de baja": "Dias_baja"
    })

    columnas = [
        c for c in [
            "Provincia",
            "Duracion_media",
            "Procesos_alta",
            "Dias_baja"
        ] if c in df.columns
    ]

    return df[columnas]


def leer_pre(archivo):
    df = pd.read_excel(archivo, header=4)
    df = df.rename(columns={
        "AMBITO GEOGRAFICO": "Provincia",
        "ÁMBITO GEOGRÁFICO": "Provincia",
        "Prevalencia por cada mil trabajadores protegidos": "Prevalencia",
        "Prevalencia": "Prevalencia",
        "Nº de procesos en vigor al final del periodo considerado": "Procesos_en_vigor",
        "Número de procesos en vigor al final del periodo considerado": "Procesos_en_vigor",
        "Trabajadores protegidos al final del periodo considerado": "Trabajadores_protegidos",
        "Trabajadores protegidos": "Trabajadores_protegidos"
    })

    columnas = [
        c for c in [
            "Provincia",
            "Prevalencia",
            "Procesos_en_vigor",
            "Trabajadores_protegidos"
        ] if c in df.columns
    ]

    return df[columnas]


dur_list = []
pre_list = []

for archivo in archivos:
    anio, organismo, contingencia, sexo, tipo = extraer_metadata(archivo.name)

    if anio is None:
        continue

    try:
        if tipo == "DUR":
            temp = leer_dur(archivo)
        elif tipo == "PRE":
            temp = leer_pre(archivo)
        else:
            continue

        temp["Año"] = anio
        temp["Organismo"] = organismo
        temp["Contingencia"] = contingencia
        temp["Sexo"] = sexo

        if tipo == "DUR":
            dur_list.append(temp)
        elif tipo == "PRE":
            pre_list.append(temp)

    except Exception as e:
        pass


dur_total = pd.concat(dur_list, ignore_index=True)
pre_total = pd.concat(pre_list, ignore_index=True)

matriz_maestra = dur_total.merge(
    pre_total,
    on=[
        "Año",
        "Provincia",
        "Organismo",
        "Contingencia",
        "Sexo"
    ],
    how="left"
)

print("DUR total:", dur_total.shape)
print("PRE total:", pre_total.shape)
print("Matriz maestra:", matriz_maestra.shape)

matriz_maestra.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

modelo_rf = RandomForestRegressor(
    n_estimators=500,
    max_depth=20,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

modelo_rf.fit(X_train, y_train)

pred_rf = modelo_rf.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

import numpy as np

mae_rf = mean_absolute_error(
    y_test,
    pred_rf
)

rmse_rf = np.sqrt(
    mean_squared_error(
        y_test,
        pred_rf
    )
)

r2_rf = r2_score(
    y_test,
    pred_rf
)

print("MAE:", mae_rf)
print("RMSE:", rmse_rf)
print("R²:", r2_rf)
```

```python
plt.figure(figsize=(6,5))

plt.scatter(
    y_test,
    pred_rf,
    alpha=0.6
)

plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    'r--'
)

plt.xlabel("Valores reales")
plt.ylabel("Predicciones")
plt.title("Random Forest: Real vs Predicho")

plt.show()
```
